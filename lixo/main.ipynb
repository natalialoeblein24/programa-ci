```python
import fitz  # PyMuPDF
import re

# Abrir o PDF
documento = fitz.open("lista.pdf")

# Definir o intervalo de páginas (19 a 32)
pagina_inicio = 19
pagina_fim = 32

# Extrair texto do intervalo de páginas
texto_completo = ""
for pagina_num in range(pagina_inicio - 1, pagina_fim):
    pagina = documento.load_page(pagina_num)
    texto = pagina.get_text("text")
    texto_completo += texto + "\n"

# Função para processar o texto e extrair dados estruturados
def processar_texto(texto):
    linhas = texto.split('\n')
    dados = []

    for linha in linhas:
        # Tentar extrair usando o formato padrão
        match = re.match(r"(\d+)\s+-\s+(.+?)\s+([\d,.]+)\s+([\d,.]+)\s+([\d,.]+)\s+([\d,.]+)", linha)
        if match:
            item = match.group(1)
            descricao = match.group(2)
            criacao = match.group(3).replace(',', '')
            finalizacao = match.group(4).replace(',', '')
            supervisao = match.group(5).replace(',', '')
            total = match.group(6).replace(',', '')
            dados.append([item, descricao, criacao, finalizacao, supervisao, total])
        else:
            # Tentar extrair usando um formato alternativo
            match = re.match(r"(\d+)\s+(.+?)\s+([\d,.]+)\s+([\d,.]+)\s+([\d,.]+)\s+([\d,.]+)", linha)
            if match:
                item = match.group(1)
                descricao = match.group(2)
                criacao = match.group(3).replace(',', '')
                finalizacao = match.group(4).replace(',', '')
                supervisao = match.group(5).replace(',', '')
                total = match.group(6).replace(',', '')
                dados.append([item, descricao, criacao, finalizacao, supervisao, total])
            else:
                # Adicionar lógica adicional para outros formatos variáveis
                pass
    
    return dados

# Processar o texto extraído
dados_extraidos = processar_texto(texto_completo)

# Imprimir os dados extraídos
for dado in dados_extraidos:
    print(dado)
```
